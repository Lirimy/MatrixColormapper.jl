```julia
function laplacian(n::Int)
    ret = SymTridiagonal(-2*ones(n), ones(n-1))
    ret[1, 1] = ret[end, end] = -1.
    ret
end

laplacian(5)
```

```julia
@time anim = openanim("rain.gif") do aio
    N = 120
    u = zeros(N, N)
    ut = zeros(u)
    Δ = laplacian(N)
    
    for i in 1:1200
        
        # rain drops
        if rand() < N^2 / 10^6
            u[rand(1:N, 2)...] = -3.
        end
        
        # simulation
        ut = ut + 0.03 * (0.5Δ*u*Δ + 2Δ*u + 2u*Δ - 0.04ut)
        u  = u + 0.03 * ut
        
        # animation output
        if i > 1000
            img = matshow((u+1)/2)
            addframe(aio, img)
        end
    end
end
```
